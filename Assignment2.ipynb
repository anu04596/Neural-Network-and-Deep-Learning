**Ques 1: What are epoch, batch size, and Stratified? Please explain them.**


Epoch:
An epoch is one complete pass of the entire training dataset through the neural network. If the dataset is trained for multiple epochs, the model learns from the data multiple times to improve performance.


Batch Size:
Batch size is the number of training samples processed before updating the model weights.

If batch size is small → faster updates, more noise
If batch size is large → stable learning, more memory usage

Stratified:
Stratified sampling ensures that the proportion of classes in the dataset is maintained during splitting (train/test). It is mainly used in classification problems to avoid class imbalance in splits.

**Ques 2: Explain the different types for activation functions?**

Activation functions introduce non-linearity into the neural network and help the model learn complex patterns.

Types of Activation Functions:

1. Sigmoid Function
Output range: 0 to 1
Used in binary classification problems.
Limitation: Vanishing gradient problem.

2. Tanh Function
Output range: -1 to 1
Better than sigmoid but still suffers from vanishing gradient.

3. ReLU (Rectified Linear Unit)
Output: f(x) = max(0, x)
Most widely used due to simplicity and efficiency.
Limitation: Dead neuron problem.

4. Leaky ReLU
Allows small negative values to overcome dead neuron issue.

5. Softmax Function
Used in multi-class classification.
Converts outputs into probability distribution.

**Ques 3: How do we handle structured and unstructured data in ANNs and CNNs?**

1. Structured Data:
Structured data is organized in tabular format such as rows and columns (e.g., CSV, Excel). Handled using Artificial Neural Networks (ANN). Requires preprocessing like encoding and scaling

2. Unstructured Data:
Unstructured data includes images, audio, and text. Handled using Convolutional Neural Networks (CNN) for images. CNN automatically extracts features from raw data

**Ques 4: What is a layer? Define ANN and CNN layers?**

Layer:
A layer is a collection of neurons that processes input data and passes output to the next layer.

1. ANN Layers:

Input Layer – Takes input features

Hidden Layers – Perform computations and learn patterns

Output Layer – Produces final prediction

2. CNN Layers:

Convolutional Layer – Extracts features using filters

Activation Layer – Applies activation function (e.g., ReLU)

Pooling Layer – Reduces dimensionality

Flatten Layer – Converts 2D data into 1D

Fully Connected Layer – Final classification

**Ques 5: What concepts are used in ANN and CNN?**

Artificial Neural Networks (ANN) and Convolutional Neural Networks (CNN) are based on several important concepts that enable them to learn from data. In ANN, key concepts include forward propagation, where input data passes through layers to produce an output; backpropagation, where errors are propagated backward to update weights; gradient descent, which is used to minimize the loss function; activation functions, which introduce non-linearity; and loss functions, which measure the difference between predicted and actual values. In CNN, along with these basic concepts, additional concepts are used such as convolution operations, where filters (kernels) are applied to extract features from input data; feature maps, which represent the output of convolution layers; stride and padding, which control how filters move across the input; and pooling operations, which reduce the spatial dimensions of the data. These concepts help CNNs effectively process and analyze unstructured data such as images
