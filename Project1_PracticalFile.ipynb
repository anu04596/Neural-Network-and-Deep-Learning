```python
#Step 1: Import Libraries
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
import numpy as np
```

```python
#Step 2: Create Dummy Dataset
X_data = np.array([[0,0], [0,1], [1,0], [1,1]], dtype=np.float32)
y_data = np.array([[0], [1], [1], [0]], dtype=np.float32)

#Step 3: Define Placeholders
X = tf.placeholder(tf.float32, shape=[None, 2])
y = tf.placeholder(tf.float32, shape=[None, 1])
```

```python
#Step 4: Initialize Weights & Biases
W1 = tf.Variable(tf.random_normal([2, 4]))
b1 = tf.Variable(tf.zeros([4]))

W2 = tf.Variable(tf.random_normal([4, 1]))
b2 = tf.Variable(tf.zeros([1]))
```

```python
#Step 5: Build Forward Propogation
# Hidden layer
z1 = tf.matmul(X, W1) + b1
a1 = tf.nn.relu(z1)

# Output layer
z2 = tf.matmul(a1, W2) + b2
output = tf.sigmoid(z2)
```

```python
#Step 6: Define Loss Function and Optimizer
loss = tf.reduce_mean(tf.square(output - y))
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.1)
train = optimizer.minimize(loss)
```

```python
#Step 7: Run Session
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())

    for epoch in range(5000):
        _, l = sess.run([train, loss], feed_dict={X: X_data, y: y_data})

        if epoch % 500 == 0:
            print("Epoch:", epoch, "Loss:", l)

    # Final predictions
    predictions = sess.run(output, feed_dict={X: X_data})
    print("Predictions:\n", predictions)
```
